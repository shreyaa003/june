```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r'C:\Users\prema\Downloads\symbols_valid_meta.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
data=data.drop_duplicates()
data.dtypes
```

```python
print(data.columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot styles
sns.set(style="whitegrid")

# 1. Missing Data Analysis
missing_data = data.isnull().sum()

# Visualize missing data
plt.figure(figsize=(10, 6))
sns.barplot(x=missing_data.index, y=missing_data.values)
plt.title('Missing Data per Column', fontsize=16)
plt.xticks(rotation=90)
plt.ylabel('Number of Missing Values')
plt.show()

# Analyze the distribution of categorical variables
categorical_columns = ['Nasdaq Traded', 'Listing Exchange', 'ETF', 'Market Category', 'Test Issue', 'Financial Status']

# Create subplots to visualize the distributions
fig, axes = plt.subplots(3, 2, figsize=(14, 14))

# Loop through and create count plots for each categorical variable
for ax, column in zip(axes.flatten(), categorical_columns):
    sns.countplot(data=data, x=column, ax=ax)
    ax.set_title(f'Distribution of {column}', fontsize=14)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

plt.tight_layout()
plt.show()

# Analyze the distribution of the numerical variable: 'Round Lot Size'
plt.figure(figsize=(8, 6))
sns.histplot(data['Round Lot Size'], bins=30, kde=True)
plt.title('Distribution of Round Lot Size', fontsize=16)
plt.xlabel('Round Lot Size')
plt.ylabel('Frequency')
plt.show()
```
